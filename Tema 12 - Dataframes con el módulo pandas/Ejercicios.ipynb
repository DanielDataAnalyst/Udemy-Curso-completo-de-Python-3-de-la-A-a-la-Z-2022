Ejercicio 6

Crea un dataframe con 10 filas y 4 columnas. La primera columna se llamará word; la segunda, length; la tercera, vowels; y la última, consonants. La columna words contendrá las siguientes 10 palabras:
“euro”, “diez” “algas”, “broma”, “cicuta”, “fatiga”, “nachos”, “jadeos”, “hazañas”, “boutique”. Las columnas length, vowels y consonants contendrán, respectivamente, la longitud, el total de vocales y el total de
consonantes. Guarda el dataframe en la variable words.

```python
def vocals(word):
  word = word.lower()
  vocals = []

  for i in word:
    if i == "a" or i == "e" or i == "i" or i == "o" or i == "u":
      vocals.append(i) 
  return len(vocals)

def cons(word):
  word = word.lower()
  cons = []

  for i in word:
    if i != "a" and i != "e" and i != "i" and i != "o" and i != "u":
      cons.append(i) 
  return len(cons)
   
import pandas as pd
words_list = ["euro", "diez", "algas", "broma", "cicuta", "fatiga", "nachos", "jadeos", "hazañas", "boutique"]
data = {"word": words_list,
        "length": map(len, words_list),
        "vowels": map(vocals, words_list),
        "consonants": map(cons, words_list)
       }

words = pd.DataFrame(data)

words
```
